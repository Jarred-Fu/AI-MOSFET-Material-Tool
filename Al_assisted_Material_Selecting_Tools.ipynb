# Section 4: 3D Visualization of the Target Region

In this section, we will visualize your MOSFET design criteria in a 3D plot. The three key properties you defined will serve as the x-, y-, and z-axes.

**Features of the Plot:**
- **Full Scale:** Each axis is displayed from 0 to **2-5 times** the target maximum value you provided.
- **Highlighted Region:** The cuboid representing your target range is highlighted with a translucent, vivid color.

Take a look at the interactive 3D plot below.


```python
import plotly.graph_objects as go
from IPython.display import display

# Check we have 3 user_properties and target_ranges
try:
    # Example structure of target_ranges:
    # [
    #   {'property': 'Property1', 'min': 10.0, 'max': None},
    #   {'property': 'Property2', 'min': None, 'max': 50.0},
    #   {'property': 'Property3', 'min': 5.0,  'max': 20.0}
    # ]
    if len(user_properties) != 3 or len(target_ranges) != 3:
        raise ValueError
except Exception:
    print("Please complete Sections 2 and 3 to define the properties and their target ranges.")
else:
    display("Section 4: 3D Visualization of the Target Region")

    # We'll compute the axis range and region boundaries for each property separately.
    axis_ranges = []
    region_mins = []
    region_maxs = []

    def compute_axis_min_max(entry):
        """Given a single target_ranges entry (dict with 'min'/'max'),
        return (axis_min, axis_max, region_min, region_max)."""
        min_val = entry.get('min', None)
        max_val = entry.get('max', None)

        # Decide on axis_min (we'll use 0 for simplicity in most cases).
        axis_min = 0.0

        # Decide on axis_max:
        # If we have a max_val, use 2×max_val. Otherwise, if we only have min_val, use 5×min_val.
        # If we have neither, default to 10.
        if max_val is not None:
            axis_max = 2 * max_val if max_val > 0 else 10
        elif min_val is not None and min_val > 0:
            axis_max = 5 * min_val
        else:
            # If everything is None or zero/negative, fallback
            axis_max = 10

        # Region boundaries:
        # region_min is min_val if that is not None, else 0
        # region_max is max_val if that is not None, else axis_max
        if min_val is not None:
            r_min = min_val
        else:
            r_min = 0  # If no min, start from 0 for the region

        if max_val is not None:
            r_max = max_val
        else:
            r_max = axis_max  # If no max, region extends up to axis max

        return (axis_min, axis_max, r_min, r_max)

    # Loop over each property to get axis range and region boundaries
    for entry in target_ranges:
        ax_min, ax_max, reg_min, reg_max = compute_axis_min_max(entry)
        axis_ranges.append([ax_min, ax_max])
        region_mins.append(reg_min)
        region_maxs.append(reg_max)

    # region_mins and region_maxs now hold the [x_min, y_min, z_min] and [x_max, y_max, z_max] for the cuboid
    x_min, y_min, z_min = region_mins
    x_max, y_max, z_max = region_maxs

    # Build vertices for the target region cuboid (Mesh3d)
    vertices = [
        [x_min, y_min, z_min],
        [x_max, y_min, z_min],
        [x_max, y_max, z_min],
        [x_min, y_max, z_min],
        [x_min, y_min, z_max],
        [x_max, y_min, z_max],
        [x_max, y_max, z_max],
        [x_min, y_max, z_max]
    ]

    # 12 triangles for the 6 faces of the cuboid
    triangles = [
        [0, 1, 2], [0, 2, 3],  # bottom face
        [4, 5, 6], [4, 6, 7],  # top face
        [0, 1, 5], [0, 5, 4],  # front face
        [2, 3, 7], [2, 7, 6],  # back face
        [0, 3, 7], [0, 7, 4],  # left face
        [1, 2, 6], [1, 6, 5]   # right face
    ]
    i_vals = [tri[0] for tri in triangles]
    j_vals = [tri[1] for tri in triangles]
    k_vals = [tri[2] for tri in triangles]

    # Create the translucent mesh for the target region
    target_region_mesh = go.Mesh3d(
        x=[v[0] for v in vertices],
        y=[v[1] for v in vertices],
        z=[v[2] for v in vertices],
        i=i_vals, j=j_vals, k=k_vals,
        color='rgba(0, 255, 0, 0.4)',
        opacity=0.4,
        name='Target Region'
    )

    # Optional empty scatter for layout context
    scatter = go.Scatter3d(x=[], y=[], z=[], mode='markers')

    fig = go.Figure(data=[scatter, target_region_mesh])

    # Label each axis with the corresponding property name
    # and set the axis range we computed.
    fig.update_layout(
        scene=dict(
            xaxis=dict(title=user_properties[0], range=axis_ranges[0]),
            yaxis=dict(title=user_properties[1], range=axis_ranges[1]),
            zaxis=dict(title=user_properties[2], range=axis_ranges[2])
        ),
        title="3D Visualization with Highlighted Target Region"
    )

    fig.show()
```

```python
def getValue(group_name="group1"):
    """
    Displays an interactive interface (Section 5 style) for defining 5 metals.
    Each property can be "Single Value" or "Range".
    The results are stored in a global dictionary `materials_data_map[group_name]`.

    Usage:
      getValue("group1")  # Data is stored in materials_data_map["group1"]
      getValue("group2")  # Data is stored in materials_data_map["group2"], etc.

    Prerequisites:
      1) user_properties is defined globally and has exactly 3 items.
      2) A global dict named materials_data_map = {} is available (created if not).
    """
    import ipywidgets as widgets
    from IPython.display import display, Markdown, clear_output

    # Make sure we have a list of three user_properties
    try:
        if len(user_properties) != 3:
            raise ValueError
    except:
        display(Markdown("**Error:** Please complete Section 2 by defining the three key properties before entering metal data."))
        return

    # If the global dictionary doesn't exist yet, create it
    global materials_data_map
    if 'materials_data_map' not in globals():
        materials_data_map = {}

    # Introduction text
    display(Markdown(f"""# {group_name}: Define Materials and Their Properties

For each material, **first** select whether each property is a *Single Value* or a *Range*.
If you pick **Single Value**, a single numeric field appears.
If you pick **Range**, two numeric fields (Min and Max) appear.

Fill in the fields for the 5 materials below, then click **Confirm Metal Data** to finalize.
"""))

    # We prompt for 5 metals
    num_metals = 5

    metal_widgets = []
    metal_boxes = []

    # Build the UI
    for i in range(num_metals):
        metal_name_widget = widgets.Text(
            value='',
            placeholder='Enter material name',
            description=f'Metal {i+1} Name:',
            style={'description_width': 'initial'}
        )

        prop_ui_list = []
        prop_vboxes = []

        for prop in user_properties:
            mode_dropdown = widgets.Dropdown(
                options=["(Select)", "Single Value", "Range"],
                value="(Select)",
                description=f"{prop}:",
                style={'description_width': 'initial'}
            )
            single_value_box = widgets.FloatText(
                description='Value:',
                style={'description_width': 'initial'}
            )
            single_value_box.layout.display = 'none'

            range_min_box = widgets.FloatText(
                description='Min:',
                style={'description_width': 'initial'}
            )
            range_max_box = widgets.FloatText(
                description='Max:',
                style={'description_width': 'initial'}
            )
            range_box = widgets.HBox([range_min_box, range_max_box])
            range_box.layout.display = 'none'

            # Toggle display logic
            def on_mode_change(change,
                               sv_box=single_value_box,
                               rng_box=range_box):
                if change['type'] == 'change' and change['name'] == 'value':
                    choice = change['new']
                    if choice == "Single Value":
                        sv_box.layout.display = ''
                        rng_box.layout.display = 'none'
                    elif choice == "Range":
                        sv_box.layout.display = 'none'
                        rng_box.layout.display = ''
                    else:
                        sv_box.layout.display = 'none'
                        rng_box.layout.display = 'none'

            mode_dropdown.observe(on_mode_change, names='value')

            prop_box = widgets.VBox([
                mode_dropdown,
                single_value_box,
                range_box
            ])

            prop_ui_list.append({
                'property': prop,
                'mode_dropdown': mode_dropdown,
                'single_value': single_value_box,
                'range_min': range_min_box,
                'range_max': range_max_box
            })
            prop_vboxes.append(prop_box)

        # Combine everything for this metal
        metal_box = widgets.VBox([
            metal_name_widget,
            widgets.HTML("<b>Properties:</b>")
        ] + prop_vboxes)

        metal_boxes.append(metal_box)
        metal_widgets.append({
            'name_widget': metal_name_widget,
            'properties_ui': prop_ui_list
        })

    # Create an accordion
    metals_input_accordion = widgets.Accordion(children=metal_boxes)
    for i in range(num_metals):
        metals_input_accordion.set_title(i, f"Metal {i+1}")

    display(metals_input_accordion)

    confirm_button = widgets.Button(
        description='Confirm Metal Data',
        button_style='success'
    )
    output_area = widgets.Output()

    def on_confirm_click(b):
        # Build the final data list
        group_data = []
        for mw in metal_widgets:
            metal_name = mw['name_widget'].value.strip()
            props = []
            for pui in mw['properties_ui']:
                prop_name = pui['property']
                mode_val = pui['mode_dropdown'].value
                if mode_val == "Single Value":
                    val = pui['single_value'].value
                    props.append({
                        'property': prop_name,
                        'mode': 'single',
                        'value': val
                    })
                elif mode_val == "Range":
                    mn = pui['range_min'].value
                    mx = pui['range_max'].value
                    props.append({
                        'property': prop_name,
                        'mode': 'range',
                        'min': mn,
                        'max': mx
                    })
                else:
                    props.append({
                        'property': prop_name,
                        'mode': None,
                        'value': None
                    })
            group_data.append({
                'material': metal_name,
                'properties': props
            })

        # Save in global dictionary
        materials_data_map[group_name] = group_data

        with output_area:
            clear_output()
            print(f"You have entered the following metal data for {group_name}:")
            for idx, mat in enumerate(group_data):
                print(f"Material {idx+1}: {mat['material']}")
                for p in mat['properties']:
                    if p['mode'] == 'single':
                        print(f"  - {p['property']}: Single Value = {p['value']}")
                    elif p['mode'] == 'range':
                        print(f"  - {p['property']}: Range = {p['min']} to {p['max']}")
                    else:
                        print(f"  - {p['property']}: [No selection made]")

    confirm_button.on_click(on_confirm_click)
    display(confirm_button, output_area)
```

# Section 6: 3D Visualization with Material Candidates and Selection

In this section, we visualize your design space along with the five material candidates you provided earlier. The 3D plot uses the three key properties as the x-, y-, and z-axes.

**Features of the Plot:**
- **Dynamic Axis Scaling:** For each axis, the range starts at 0 and extends to the greater of (2 × target maximum) or the highest material candidate value.
- **Highlighted Region:** The target region is displayed as a translucent green cuboid.
- **Material Candidates:** Your five material candidates are shown as red markers, with their names labeled.

Based on the graph, please select the material that best meets your design criteria.


```python
def getGraph(group_name="group1"):
    """
    Visualizes the data for the given group_name in a 3D plot and
    allows the user to select the best performing material.

    Requirements:
      1) user_properties (length 3) and target_ranges defined globally.
      2) materials_data_map[group_name] contains a list of 5 material dicts
         with "mode": 'single' or 'range', etc.
      3) final_best_material_map is a global dict to store the chosen best material.

    Usage:
      getGraph("group1")
        -> Reads from materials_data_map["group1"]
        -> Displays the 3D plot
        -> Lets user pick the best material
        -> Stores in final_best_material_map["group1"]
    """
    import plotly.graph_objects as go
    import ipywidgets as widgets
    from IPython.display import display, Markdown, clear_output

    # Check that the needed globals exist
    try:
        # Ensure user_properties and target_ranges
        if len(user_properties) != 3 or len(target_ranges) != 3:
            raise ValueError("user_properties or target_ranges not properly defined.")
        # Ensure the group data is present
        if group_name not in materials_data_map:
            raise KeyError(f"No data found for {group_name} in materials_data_map.")
        if len(materials_data_map[group_name]) != 5:
            raise ValueError(f"{group_name} data incomplete or not 5 materials.")
    except Exception as e:
        display(Markdown(f"**Error:** {str(e)}"))
        return

    # We'll reference the group's material data as 'materials'
    materials = materials_data_map[group_name]

    # Display a header
    display(Markdown(f"## 3D Visualization for {group_name}"))

    # Helper function: get (min, max) from a single property dict
    def get_prop_min_max(prop_dict):
        mode = prop_dict.get('mode')
        if mode == 'single':
            val = prop_dict.get('value', 0.0)
            return (val, val)
        elif mode == 'range':
            mn = prop_dict.get('min', 0.0)
            mx = prop_dict.get('max', mn)
            return (mn, mx)
        return (0.0, 0.0)

    # 1) Determine axis bounds from target_ranges + these materials
    axis_mins = [None, None, None]
    axis_maxs = [None, None, None]

    # Incorporate target ranges
    for i, tdict in enumerate(target_ranges):
        tmin = tdict.get('min')
        tmax = tdict.get('max')
        if tmin is not None:
            if axis_mins[i] is None or tmin < axis_mins[i]:
                axis_mins[i] = tmin
        if tmax is not None:
            if axis_maxs[i] is None or tmax > axis_maxs[i]:
                axis_maxs[i] = tmax

    # Incorporate all materials
    for mat in materials:
        for i, pdict in enumerate(mat['properties']):
            mn, mx = get_prop_min_max(pdict)
            if axis_mins[i] is None or mn < axis_mins[i]:
                axis_mins[i] = mn
            if axis_maxs[i] is None or mx > axis_maxs[i]:
                axis_maxs[i] = mx

    # Handle fallback if still None or min==max
    for i in range(3):
        if axis_mins[i] is None:
            axis_mins[i] = 0.0
        if axis_maxs[i] is None or axis_maxs[i] == axis_mins[i]:
            axis_maxs[i] = axis_mins[i] + 10.0
        # If min is positive, clamp to 0
        if axis_mins[i] > 0:
            axis_mins[i] = 0.0
        axis_maxs[i] *= 1.2  # Expand by 20%

    # Helper to build a cuboid
    def build_cuboid(x_min, x_max, y_min, y_max, z_min, z_max):
        vertices = [
            [x_min, y_min, z_min],
            [x_max, y_min, z_min],
            [x_max, y_max, z_min],
            [x_min, y_max, z_min],
            [x_min, y_min, z_max],
            [x_max, y_min, z_max],
            [x_max, y_max, z_max],
            [x_min, y_max, z_max]
        ]
        faces = [
            [0,1,2],[0,2,3],   # bottom
            [4,5,6],[4,6,7],   # top
            [0,1,5],[0,5,4],   # front
            [2,3,7],[2,7,6],   # back
            [0,3,7],[0,7,4],   # left
            [1,2,6],[1,6,5]    # right
        ]
        i_vals = [f[0] for f in faces]
        j_vals = [f[1] for f in faces]
        k_vals = [f[2] for f in faces]
        return vertices, i_vals, j_vals, k_vals

    # 2) Build Target Region (Mesh3d)
    def get_dim_min_max(tdict, i):
        tmin = tdict.get('min')
        tmax = tdict.get('max')
        amn = axis_mins[i]
        amx = axis_maxs[i]
        mn = tmin if (tmin is not None) else amn
        mx = tmax if (tmax is not None) else amx
        return (mn, mx)

    x_min_t, x_max_t = get_dim_min_max(target_ranges[0], 0)
    y_min_t, y_max_t = get_dim_min_max(target_ranges[1], 1)
    z_min_t, z_max_t = get_dim_min_max(target_ranges[2], 2)
    tv, ti, tj, tk = build_cuboid(x_min_t, x_max_t, y_min_t, y_max_t, z_min_t, z_max_t)
    target_region = go.Mesh3d(
        x=[v[0] for v in tv],
        y=[v[1] for v in tv],
        z=[v[2] for v in tv],
        i=ti, j=tj, k=tk,
        color='rgba(0,255,0,0.4)',
        opacity=0.4,
        name='Target Region'
    )

    # 3) Visualize each material
    # If all 3 properties are single => one 3D point
    # Otherwise => Mesh3D bounding region
    def all_single(props):
        return all((p.get('mode') == 'single') for p in props)

    colors = [
        'rgba(255,0,0,0.4)',
        'rgba(0,0,255,0.4)',
        'rgba(255,165,0,0.4)',
        'rgba(128,0,128,0.4)',
        'rgba(0,255,255,0.4)',
        'rgba(255,0,255,0.4)',
        'rgba(0,128,0,0.4)',
    ]
    fig_data = [target_region]

    for idx, mat in enumerate(materials):
        mat_name = mat['material']
        props = mat['properties']
        mat_color = colors[idx % len(colors)]
        if all_single(props):
            # single point
            x_val = props[0].get('value', 0.0)
            y_val = props[1].get('value', 0.0)
            z_val = props[2].get('value', 0.0)
            single_scatter = go.Scatter3d(
                x=[x_val],
                y=[y_val],
                z=[z_val],
                mode='markers+text',
                marker=dict(size=5, color=mat_color),
                text=[mat_name],
                textposition='top center',
                name=mat_name
            )
            fig_data.append(single_scatter)
        else:
            # Build bounding cuboid
            px_min, px_max = get_prop_min_max(props[0])
            py_min, py_max = get_prop_min_max(props[1])
            pz_min, pz_max = get_prop_min_max(props[2])
            mv, mi, mj, mk = build_cuboid(px_min, px_max, py_min, py_max, pz_min, pz_max)
            mat_mesh = go.Mesh3d(
                x=[v[0] for v in mv],
                y=[v[1] for v in mv],
                z=[v[2] for v in mv],
                i=mi, j=mj, k=mk,
                color=mat_color,
                opacity=0.5,
                name=mat_name
            )
            fig_data.append(mat_mesh)

            # Center label
            cx = (px_min + px_max)/2
            cy = (py_min + py_max)/2
            cz = (pz_min + pz_max)/2
            center_label = go.Scatter3d(
                x=[cx], y=[cy], z=[cz],
                mode='text+markers',
                text=[mat_name],
                textposition='top center',
                marker=dict(size=5, color=mat_color),
                name=f"{mat_name} center"
            )
            fig_data.append(center_label)

    # Build and show the figure
    fig = go.Figure(data=fig_data)
    fig.update_layout(
        scene=dict(
            xaxis=dict(title=user_properties[0], range=[axis_mins[0], axis_maxs[0]]),
            yaxis=dict(title=user_properties[1], range=[axis_mins[1], axis_maxs[1]]),
            zaxis=dict(title=user_properties[2], range=[axis_mins[2], axis_maxs[2]])
        ),
        title=f"3D Visualization for {group_name}"
    )
    fig.show()

    # 4) Select the best material
    display(Markdown(f"### Choose the Best Material Candidate for {group_name}"))
    mat_names = [m['material'] for m in materials]
    material_dropdown = widgets.Dropdown(
        options=mat_names,
        description='Best Material:',
        style={'description_width': 'initial'}
    )
    submit_button = widgets.Button(description="Submit Selection", button_style='success')
    selection_output = widgets.Output()

    # We'll store the best selection in final_best_material_map[group_name]
    global final_best_material_map
    if 'final_best_material_map' not in globals():
        final_best_material_map = {}

    def on_submit_clicked(b):
        selected = material_dropdown.value
        chosen_metal = None
        for cmat in materials:
            if cmat['material'] == selected:
                chosen_metal = cmat
                break
        final_best_material_map[group_name] = chosen_metal
        with selection_output:
            clear_output()
            print(f"Submitted best material for {group_name}: {chosen_metal}")

    submit_button.on_click(on_submit_clicked)
    display(material_dropdown, submit_button, selection_output)
```

# Section 8: Final Metal Selection

Now that you have completed all 3 iterations, the best metal from each round has been recorded. The final 3D plot below shows these 5 best metals (with dynamic axis scaling and the target region). Based on this visualization, please select the final best metal candidate.


```python
def finalSection8():
    """
    Final step: Retrieve the best metals from each group (group1..group5)
    and plot them in 3D with the target region. Then let the user pick the
    absolute best among those 5.

    Requirements:
      - final_best_material_map must contain entries for 'group1'..'group5'
        (like final_best_material_map['group1'] = {...})
      - user_properties and target_ranges as in earlier sections
      - We store the final selection in a global variable: final_best_overall
    """
    import plotly.graph_objects as go
    import ipywidgets as widgets
    from IPython.display import display, clear_output, Markdown

    # Check that we have user_properties and target_ranges
    try:
        if len(user_properties) != 3 or len(target_ranges) != 3:
            raise ValueError("user_properties or target_ranges not valid.")
    except Exception as e:
        display(Markdown(f"**Error:** {str(e)}"))
        return

    # We expect final_best_material_map with 'group1'..'group5'
    groups = ["group1", "group2", "group3", "group4", "group5"]
    # groups = ["group1", "group2"]
    best_mats = []
    missing = []
    for g in groups:
        if g in final_best_material_map and final_best_material_map[g] is not None:
            best_mats.append((g, final_best_material_map[g]))
        else:
            missing.append(g)
    if missing:
        display(Markdown(f"**Error:** No best metal found for {', '.join(missing)}. Please complete those first."))
        return

    # Now best_mats is a list of (groupName, materialDict)
    # We'll create a 3D plot for these 5 best materials + target region.

    display(Markdown("## Final Section 8: Compare the 5 Best Metals"))

    # Helper to read min/max from a property dict
    def get_prop_min_max(prop_dict):
        mode = prop_dict.get('mode')
        if mode == 'single':
            val = prop_dict.get('value', 0.0)
            return (val, val)
        elif mode == 'range':
            mn = prop_dict.get('min', 0.0)
            mx = prop_dict.get('max', mn)
            return (mn, mx)
        return (0.0, 0.0)

    # We'll compute global axis ranges from target_ranges + these 5 metals
    axis_mins = [None, None, None]
    axis_maxs = [None, None, None]

    # Incorporate target ranges
    for i, tdict in enumerate(target_ranges):
        tmin = tdict.get('min')
        tmax = tdict.get('max')
        if tmin is not None:
            if axis_mins[i] is None or tmin < axis_mins[i]:
                axis_mins[i] = tmin
        if tmax is not None:
            if axis_maxs[i] is None or tmax > axis_maxs[i]:
                axis_maxs[i] = tmax

    # Incorporate the 5 best materials
    for (g, mat) in best_mats:
        props = mat['properties']
        for i, pdict in enumerate(props):
            mn, mx = get_prop_min_max(pdict)
            if axis_mins[i] is None or mn < axis_mins[i]:
                axis_mins[i] = mn
            if axis_maxs[i] is None or mx > axis_maxs[i]:
                axis_maxs[i] = mx

    # Fallback if still None or min==max
    for i in range(3):
        if axis_mins[i] is None:
            axis_mins[i] = 0.0
        if axis_maxs[i] is None or axis_maxs[i] == axis_mins[i]:
            axis_maxs[i] = axis_mins[i] + 10.0
        if axis_mins[i] > 0:
            axis_mins[i] = 0.0
        axis_maxs[i] *= 1.2

    # Helper to build a cuboid
    def build_cuboid(x_min, x_max, y_min, y_max, z_min, z_max):
        vertices = [
            [x_min, y_min, z_min],
            [x_max, y_min, z_min],
            [x_max, y_max, z_min],
            [x_min, y_max, z_min],
            [x_min, y_min, z_max],
            [x_max, y_min, z_max],
            [x_max, y_max, z_max],
            [x_min, y_max, z_max]
        ]
        faces = [
            [0,1,2],[0,2,3],   # bottom
            [4,5,6],[4,6,7],   # top
            [0,1,5],[0,5,4],   # front
            [2,3,7],[2,7,6],   # back
            [0,3,7],[0,7,4],   # left
            [1,2,6],[1,6,5]    # right
        ]
        i_vals = [f[0] for f in faces]
        j_vals = [f[1] for f in faces]
        k_vals = [f[2] for f in faces]
        return vertices, i_vals, j_vals, k_vals

    # Build target region (Mesh3d)
    def get_dim_min_max(tdict, i):
        tmin = tdict.get('min')
        tmax = tdict.get('max')
        amn = axis_mins[i]
        amx = axis_maxs[i]
        mn = tmin if (tmin is not None) else amn
        mx = tmax if (tmax is not None) else amx
        return (mn, mx)

    x_min_t, x_max_t = get_dim_min_max(target_ranges[0], 0)
    y_min_t, y_max_t = get_dim_min_max(target_ranges[1], 1)
    z_min_t, z_max_t = get_dim_min_max(target_ranges[2], 2)
    tv, ti, tj, tk = build_cuboid(x_min_t, x_max_t, y_min_t, y_max_t, z_min_t, z_max_t)
    target_region = go.Mesh3d(
        x=[v[0] for v in tv],
        y=[v[1] for v in tv],
        z=[v[2] for v in tv],
        i=ti, j=tj, k=tk,
        color='rgba(0,255,0,0.4)',
        opacity=0.4,
        name='Target Region'
    )

    # Different colors for the 5 best
    colors = [
        'rgba(255,0,0,0.4)',
        'rgba(0,0,255,0.4)',
        'rgba(255,165,0,0.4)',
        'rgba(128,0,128,0.4)',
        'rgba(0,255,255,0.4)'
    ]

    fig_data = [target_region]

    def all_single(props):
        return all(p.get('mode') == 'single' for p in props)

    # Plot each best metal
    for idx, (g, mat) in enumerate(best_mats):
        mat_name = f"{g}: {mat['material']}"
        props = mat['properties']
        mat_color = colors[idx % len(colors)]
        if all_single(props):
            # single point
            x_val = props[0].get('value', 0.0)
            y_val = props[1].get('value', 0.0)
            z_val = props[2].get('value', 0.0)
            scatter = go.Scatter3d(
                x=[x_val],
                y=[y_val],
                z=[z_val],
                mode='markers+text',
                marker=dict(size=5, color=mat_color),
                text=[mat_name],
                textposition='top center',
                name=mat_name
            )
            fig_data.append(scatter)
        else:
            # bounding cuboid
            px_min, px_max = get_prop_min_max(props[0])
            py_min, py_max = get_prop_min_max(props[1])
            pz_min, pz_max = get_prop_min_max(props[2])
            mv, mi, mj, mk = build_cuboid(px_min, px_max, py_min, py_max, pz_min, pz_max)
            mat_mesh = go.Mesh3d(
                x=[v[0] for v in mv],
                y=[v[1] for v in mv],
                z=[v[2] for v in mv],
                i=mi, j=mj, k=mk,
                color=mat_color,
                opacity=0.5,
                name=mat_name
            )
            fig_data.append(mat_mesh)
            # center label
            cx = (px_min + px_max)/2
            cy = (py_min + py_max)/2
            cz = (pz_min + pz_max)/2
            center_label = go.Scatter3d(
                x=[cx], y=[cy], z=[cz],
                mode='text+markers',
                text=[mat_name],
                textposition='top center',
                marker=dict(size=5, color=mat_color),
                name=mat_name
            )
            fig_data.append(center_label)

    fig = go.Figure(data=fig_data)
    fig.update_layout(
        scene=dict(
            xaxis=dict(title=user_properties[0], range=[axis_mins[0], axis_maxs[0]]),
            yaxis=dict(title=user_properties[1], range=[axis_mins[1], axis_maxs[1]]),
            zaxis=dict(title=user_properties[2], range=[axis_mins[2], axis_maxs[2]])
        ),
        title="Final 3D Comparison of the 5 Best Metals"
    )
    fig.show()

    # Prompt user to pick the best of these 5
    display(Markdown("### Select the Final Best Metal Among These 5"))
    best_names = [f"{g}: {m['material']}" for g, m in best_mats]
    final_dropdown = widgets.Dropdown(
        options=best_names,
        description='Final Best:',
        style={'description_width': 'initial'}
    )
    submit_button = widgets.Button(description="Submit Final Best", button_style='success')
    selection_output = widgets.Output()

    global final_best_overall
    final_best_overall = None

    def on_submit_final(b):
        selected = final_dropdown.value
        # Find the matching group, metal
        chosen_group = None
        chosen_mat = None
        for (g, m) in best_mats:
            label = f"{g}: {m['material']}"
            if label == selected:
                chosen_group = g
                chosen_mat = m
                break
        with selection_output:
            clear_output()
            print("Final Best Selected:", chosen_group, chosen_mat)
        # Store in a global
        final_best_overall = {
            'group': chosen_group,
            'material': chosen_mat
        }

    submit_button.on_click(on_submit_final)
    display(final_dropdown, submit_button, selection_output)
```
